**Reduce las dimensiones con PCA, si consideras necesario.
Indica la varianza de los datos explicada por cada componente seleccionado. Para actividades de exploración de los datos la varianza > 70%
Indica la importancia de las variables en cada componente**

```python
#librería para aplicar PCA
from sklearn.decomposition import PCA
#sacamos PCA
pca=PCA()
pca_v = pca.fit_transform(scaled_paydf)
#mostramos los primeros 10 valores
pca_v[:10]
```

```python
#obtener varianza explicada por componente y la varianza acumulada
pcaObj_paydf = pd.DataFrame({
'Varianza explicada%': np.round(pca.explained_variance_ratio_,4) * 100,
'Varianza acumulada%': np.cumsum(pca.explained_variance_ratio_) * 100})
pcaObj_paydf
```

```python
#añadimos nombres a las filas para un análisis mas sencillo 
pca_names = [f'Component{i + 1}' for i in range(len(scaled_paydf.columns))]
pcaObj_paydf.index = pca_names
pcaObj_paydf
```

```python
#declaración de dataframe con pesos de componentes
pca_df = pd.DataFrame(pca_v, columns =pca_names)
#comparamos varianzas de variables y componentes
print("Varianza de variables originales: ", scaled_paydf.var().sum())
print("Varianza de componentes: ", pca_df.var().sum())
```

```python
#análisis para encontrar los componentes principales
pcaObj_paydf
#viendo la columna de varianza acumulada podemos concluir que nuestros componentes principales pueden ser de los primeros 4 a 5 componentes, ya que con 5 la variación se ve explicada en un 95%
```

```python
#comparación con variables originales escaladas para alcanzar el 95% de variación
total =scaled_paydf.var().sum()
pd.DataFrame({
"Varianza %": (scaled_paydf.var()/ total) * 100,
"Varianza acumulada %": (scaled_paydf.var().cumsum() / total) *100
})
#podemos notar que sin sacar el PCA necesitaríamos utilizar 10/11 variables en vez de 5 para llegar a más del 90% de la variación explicada, se logró reducir a la mitad de factores
```

```python
#analizamos la composición de los 5 componentes principales que contienen la mayoría de información
comp_paydf = pd.DataFrame(
pca.components_.round(2), 
columns = pca_df.columns, 
index = scaled_paydf.columns) 
comp_paydf.iloc[:,:5]
```

```python
#obtener los valores máximo por componente en filas para analizar mejor la información
comp_paydf.iloc[:,:10].abs().idxmax()
#mostrando los 10 primeros componentes podemos ver que la misma variable puede ser la que más contribuya para distintos componentes, en este caso julybillstatement y junebillstatement son las que mas contribuyen en varios componentes
```

**Interpreta y explica cada uno de los gráficos indicando cuál es la información más relevante que podría ayudar en el proceso de toma de decisiones.**

El objetivo de realizar un análisis de componentes principales (PCA, por sus siglas en inglés). Este se utiliza para disminuir el tamaño de una base de datos, y por lo tanto, poder realizar análisis de bases de datos muy grandes, de forma más eficiente, pero manteniendo la mayor cantidad de información posible. Partiendo de este objetivo, podemos determinar que no existe una respuesta “correcta o incorrecta” a la pregunta, sino que la respuesta dependerá del criterio que el analista determine, en base a qué tan eficiente quiere hacer el análisis, y qué tanta información está dispuesto o no a sacrificar. Consideramos que, en el caso del análisis de esta base de datos sobre comportamientos crediticios, un número mínimo adecuado sería el uso de 5 componentes. 
 
Con cinco componentes, seguimos utilizando menos de la mitad de los componentes (si partimos de que contamos con once componentes), pero a la vez logramos mantener hasta el 95% de la información presentada por la base de datos. Tal vez podríamos utilizar incluso dos componentes menos (considerando la Ley de Pareto, podríamos considerar satisfactorio un mínimo del 80% de la información), pero preferimos sacrificar aún menos información.
 

La pérdida de información se da de acuerdo a la varianza acumulada de los componentes 6 al 11; es decir, la varianza restante que no está incluida en los componentes elegidos 1 al 5. En este caso, la pérdida de información corresponde al 4.99% del total de la información presentada por las 11 variables numéricas de la base de datos.
 

Mostrando los 10 primeros componentes podemos ver que la misma variable puede ser la que más contribuya para distintos componentes, en este caso JulyBillStatement y JuneBillStatement son las que más contribuyen en varios componentes. La variable de mayor importancia es JulyBillStatement, la variable de menor importancia es totalBillStatement.
 

Un análisis de componentes principales (PCA, por sus siglas en inglés) se recomienda realizar especialmente cuando estamos tratando con bases de datos muy grandes, ya que el beneficio principal de este análisis es el de reducir el tamaño de la base de datos para poder simplificar el desarrollo de un modelo de machine learning, y en análisis y visualización de los datos, pero perdiendo el mínimo de información. En resumen, nos permite reducir el número de variables al mínimo, al tiempo que conservamos la mayor información posible.

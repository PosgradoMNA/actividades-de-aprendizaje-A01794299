# **Working with Numpy Arrays & Simple APIs**

**Numby 1D Arrays**

Es una librería para la computación científica. Es la base para los pandas.

**The basics and array creation**

Un arreglo numpy o ndarray es similar a una lista, usualmente tiene un tamaño justo y cada elemento es del mismo tipo.

Documentación de la librería:

[numpy.org](https://numpy.org)



**Cast list to numpy array:**

```python
import numpy as np
a=np.array([0,1,2,3,4])
a
#acceder a un index específico
a[0]
#revisar tipo
type(a)
#como el array tiene datos del mismo tipo, se puede usar dtype para obtener el tipo de dato de los elementos del array
a.dtype
#tamaño del array
a.size
#numero de dimensiones en el arreglo o el rango del arreglo
a.ndim
#tupla de ints que indica el tamaño del arreglo en cada dimensión
a.shape
```

**Indexing and slicing**


```python
#Cambiar el primer elemento del array a 100
c=np.array([20,1,2,3,4])
c[0]=100
c
#slicing de elementos del index 1 al 4, como en las listas, en el resultado no se cuenta el elemento del último index
d=c[1:4]
d
#asignar nuevos valores a varios index
c[3:5]=300,400
c
```

**Basic operations**

Las mismas operaciones son usualmente mas rápidas y requieren menos memoria en numpy comparado con python regular.

**Vector addition and substraction**
image.png

```python
#primera forma de sumar vectores en python, unir las dos listas y guardar los resultados en z
u=[1,0]
v=[1,0]
z=[]
for n, m in zip(u,v):
  z.append(n+m)
z
#segunda forma de suma de vectores con numpy
u=np.array([1,0])
v=np.array([0,1])
z=u+v
z
#substracción (resta) de vectores
z=u-v
z
```

**Array multiplication with Scalar**


```python
#multiplicación de vector por scalar 2 en numpy
y=np.array([1,2])
z=2*y
z
#multiplicación de vector en python regular (mas lento)
y=[1,2]
z=[]

for n in y:
  z.append(2*n)

z
```

**Hadamard product of two numpy arrays**

```python
#Producto de dos vectores (Hadamard product)
u=np.array([1,2])
v=np.array([3,2])
z=u*v
z
```

**Dot Product**

```python
#Suma de la multiplicación de dos vectores
#numpy dot function
result=np.dot(u,v)
result
```

**Adding constant to an numpy Array**

```python
#sumar un valor constante a todos los elementos del array
u=np.array([1,2,3,-1])
z=u+1
z
```

# **2-Dimensional Numpy Arrays**

Se pueden crear arreglos con mas de una dimensión.



```python
#lista con tres listas anidadas del mismo tamaño
a=[[11,12,13],[21,22,23],[31,32,33]]
#cambiar a numpy array
A=np.array(a)
A
```

```python
#ndim para obtener las dimensiones o axis
A.ndim
#el número de listas no tiene que ver con las dimensiones, sino con la forma de la lista
#shape regresa una tupla
A.shape
#tamaño de array
A.size
#acceder a un index específico
A[1][0]
#slicing
#axis primera linea, primeras dos columnas
A[0,0:2]
#primer index corresponde a las ultimas dos lineas, segundo index accesa la ultima columna
A[0:2,2]
#agregar arrays
X=np.array([[1,0],[0,1]])
Y=np.array([[2,1],[1,2]])
Z=X+Y
Z
```

```python
#multiplicar por scalar
Z=2*Y
Z
#multiplicación de dos arreglos Hadamard
Z=X*Y
Z
```

**Matrix multiplication**

El número de columnas de la primera matriz debe ser el mismo que el número de líneas de la segunda matriz


```python
A=np.array([[0,1,1],[1,0,1]])
B=np.array([[1,1],[1,1],[-1,1]])
#sacar el producto punto de A y B
C=np.dot(A,B)
C
```

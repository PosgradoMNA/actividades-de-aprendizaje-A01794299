# **Problema: Evaluación de carro usado**

¿Podemos estimar el precio de los carros usados?

```python
#librerías
import pandas as pd
import numpy as np
```

```python
#importar dataset
path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
```

```python
#convertir dataset en csv, especificando que las columnas no tienen nombres específicos
df = pd.read_csv(path, header=None)
#primeras 5 filas del dataset
df.head()
```

```python
#últimas 5 filas del dataframe
df.tail(5)
```

```python
#especificamos los nombres de las columnas (headers)
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
#asignamos los nombres al dataset
df.columns = headers
df.head()
```

```python
#eliminamos los valores ? declarandolos como NaN
dfnan=df.replace('?',np.NaN)
dfnan
```

```python
#eliminamos los datos NaN en base a la columna normalized-losses del dataframe
df=dfnan.dropna(subset=['normalized-losses'],axis=0)
df.head(10)
```

```python
#guardamos el archivo como csv, sin sus índices
df.to_csv("automobile.csv", index=False)
```

```python
#revisar los tipos de dato en pandas
df.dtypes
```

**Missing values**

¿Qué es un valor faltante?

Ocurren cuando ningun valor de dato es almacenado para una variable en una observación.

Puede representarse como "?", "N/A", 0 o una celda vacía.

¿Cómo lidiar con datos vacíos?

-Revisar con la fuente o persona que recolectó los datos y ver si pueden encontrar el valor actual.
-Remover el dato donde se encuentra el valor vacío.
-Cuando se remueven datos, puede removerse la variable completa o solo la entrada de datos con el valor vacío, si no hay muchas observaciones con datos vacíos, usualmente es mejor remover la entrada particular.
-Reemplazar datos es mejor, ya que ningún dato es desperdiciado, pero es menos preciso porque se deben reemplazar los datos con un aproximado de lo que el dato debería ser, una técnica de reemplazo estandar es cambiar el valor vacío por el valor promedio de la variable total.
-Reemplazo basado en otras funciones.
-Dejar los valores vacíos como están.

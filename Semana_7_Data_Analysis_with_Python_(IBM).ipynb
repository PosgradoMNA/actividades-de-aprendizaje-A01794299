# **Linear Regression and Multiple Linear Regression**

-La regresión linear usará una variable independiente para hacer una predicción.
-La regresión linear múltiple se referirá a multiples variables independientes para hacer la predicción.

**Simple Linear Regression**

image.png

```python
from sklearn.linear_model import LinearRegression
```

```python
import pandas as pd
import numpy as np
#importar dataset
path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
```

```python
#convertir dataset en csv, especificando que las columnas no tienen nombres especÃƒÂ­ficos
df = pd.read_csv(path, header=None)
#primeras 5 filas del dataset
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
#asignamos los nombres al dataset
df.columns = headers
df=df.replace('?',np.NaN)
df=df.dropna(subset=['normalized-losses'],axis=0)
df.head()
```

```python
#crear objeto de regresión linear usando el constructor
lm=LinearRegression()
```

```python
#definir la variable predictora y la variable objetivo
X= df[['highway-mpg']]
Y= df['price']
#usamos fit para ajustar el modelo
lm.fit(X,Y)
#obtenemos la predicción
Yhat=lm.predict(X)
Yhat
```

```python
#b0 view the intercept
lm.intercept_
```

```python
#b1 view the slope
lm.coef_
```

```python
#relación entre precio y highway mpg está dada por:
price= lm.intercept_ - lm.coef_ * highway-mpg
```

**Multiple Linear Regression (MLR)**

Este método es utilizado para explicar la relación entre una variable objetivo continua y dos o más variables predictoras.

**Fitting a multiple linear model estimator**

```python
#extraer las 4 variables predictoras y guardarlas en Z
Z= df[['horsepower','curb-weight','engine-size','highway-mpg']]
#entrenar el modelo
lm.fit(Z,df['price'])
#obtener la predicción
Yhat=lm.predict(Z)
Yhat
```

¿Por qué usar gráficas de regresión?

Nos da una buena estimación de:

- La relación entre dos variables
- La fuerza de la correlación
- La dirección de la relación (positiva o negativa)
